```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer_churn_dataset.csv')
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
df.head()
```

```python
le=LabelEncoder()
gender=le.fit_transform(df['Gender'])

df['Gender']=gender
```

```python
ohe=OneHotEncoder(sparse_output=False)


contract_length=ohe.fit_transform(df[['Contract Length']])

contract_length=pd.DataFrame(contract_length,columns=ohe.get_feature_names_out())

df=pd.concat([df,contract_length],axis=1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
print(df['Subscription Type'].unique())
```

```python
subscription=ohe.fit_transform(df[['Subscription Type']])

subscription=pd.DataFrame(subscription,columns=ohe.get_feature_names_out())

df=pd.concat([df,subscription],axis=1)
```

```python
df.head()
```

```python
new_df=df.drop(columns=['Contract Length','Subscription Type'])
```

```python

new_df.head()
```

```python
new_df['Churn'].value_counts()
```

```python
new_df.corr()['Churn'].sort_values(ascending=False)
```

```python
new_df.dropna(inplace=True)
```

```python
new_df['Churn'].astype(int)
```

```python
df['Churn'].dropna(inplace=True)
```

```python
X=new_df.drop(columns=['Churn','CustomerID'])
y=new_df['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=25)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
y_train_pred=rfc.predict(X_train)
y_test_pred=rfc.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
#random forest
roc_score={}
probs=rfc.predict_proba(X_test)[:,1]
roc_score['random_forest']=roc_auc_score(y_test,probs)


```

```python
from sklearn.metrics import classification_report , confusion_matrix

report=classification_report(y_train,y_train_pred)
matrix=confusion_matrix(y_train,y_train_pred)
print(report)
print(matrix)


report=classification_report(y_test,y_test_pred)
matrix=confusion_matrix(y_test,y_test_pred)
print(report)
print(matrix)
```

```python
from xgboost import XGBClassifier
model_xgb=XGBClassifier()
model_xgb.fit(X_train,y_train)
y_train_pred=model_xgb.predict(X_train)
y_test_pred=model_xgb.predict(X_test)
```

```python
probs=model_xgb.predict_proba(X_test)[:,1]
roc_score['xgboost']=roc_auc_score(y_test,probs)
```

```python
report=classification_report(y_train,y_train_pred)
matrix=confusion_matrix(y_train,y_train_pred)
print(report)
print(matrix)


report=classification_report(y_test,y_test_pred)
matrix=confusion_matrix(y_test,y_test_pred)
print(report)
print(matrix)
```

```python
from sklearn.linear_model import LogisticRegression

regressor=LogisticRegression(max_iter=1000)
regressor.fit(X_train,y_train)
y_train_pred=regressor.predict(X_train)
y_test_pred=regressor.predict(X_test)
```

```python
probs=regressor.predict_proba(X_test)[:,1]
roc_score['logisticRegression']=roc_auc_score(y_test,probs)
```

```python
report=classification_report(y_train,y_train_pred)
matrix=confusion_matrix(y_train,y_train_pred)
print(report)
print(matrix)


report=classification_report(y_test,y_test_pred)
matrix=confusion_matrix(y_test,y_test_pred)
print(report)
print(matrix)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5,algorithm='auto',weights='uniform')
knn.fit(X_train,y_train)
y_train_pred=knn.predict(X_train)
y_test_pred=knn.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB

nb=GaussianNB()
nb.fit(X_train,y_train)
y_train_pred=nb.predict(X_train)
y_test_pred=nb.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_train_pred=gb.predict(X_train)
y_test_pred=gb.predict(X_test)
```

```python
probs=gb.predict_proba(X_test)[:,1]
roc_score['gradientBoosting']=roc_auc_score(y_test,probs)
```

```python
for i in roc_score:
    print(i,roc_score[i])
```

```python
models={
    'random_forest':RandomForestClassifier(),
    'xgboost':XGBClassifier(),
    'logisticRegression':LogisticRegression(max_iter=1000),
    'knn':KNeighborsClassifier(n_neighbors=5,algorithm='auto',weights='uniform'),
    'naive_bayes':GaussianNB(),
    'gradientBoosting':GradientBoostingClassifier()
}


best_model_name=max(roc_score,key=roc_score.get)
for name in models:
    if name==best_model_name:
        best_model=models[name]
print(best_model_name)
print(best_model)
```
